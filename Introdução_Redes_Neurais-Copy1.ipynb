$$Conceitos:$$

- **Camadas de entrada não pode ser menor do que a quantidade de variáveis que estamos usando para explicar Y**


- **Deep Learning** = Arquitetura da Rede Neural


- **Redes neurais rasas (shallow)** = Possui apenas uma camada oculta


- **Redes neurais profundas** = Quando as redes neurais possuem muitas camadas ocultas, trata-se de deep learning.


- **Função de ativação** = 


- **Função de Custo** = Função que mensura o quanto estamos acertando


- **Epocas** =


- **Descida do gradiente** = Processo utilizado para encontrar o mínimo da função custo com base em variação nos pesos, através do processo de backpropagation os erros obtidos após os cálculos dos pesos são retroalimentados na rede até otimizarmos o resultado


**Tipos de Modelos de Redes Neurais:**


- **Perceptron (P), Feed Forward Network (FFN), Radial Basis Network (RBF)** = São o tipo mais básico, onde a informação de entrada flui em sequência linear até a saída.

```python
#Separar os dados em X e Y
x = dados.iloc[:,0:11]
y = dados.quality

#Padronização dos dados

dados_padro = (x - x.mean()) / (x.std())
dados_padro
```

```python
#Divisão dos dados 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dados_padro,y,test_size=0.3,random_state=1)

#Alterar as dimensoes
X_train = X_train.T
X_test = X_test.T
y_train = pd.DataFrame(y_train).T
y_test = pd.DataFrame(y_test).T
```

```python
#Função sigmoid    
def sig(x):
    return((1/(1+np.exp(-x))))
```

```python
class Total:    
#Iniciar os parametros da rede    
    def __init__(self,x,y,hidden_neurons,lr,num_iteration):
        self.x = x
        self.y = y
        self.hidden_neurons = hidden_neurons
        self.lr = lr
        self.num_iteration = num_iteration
        
        
#Arquitetura da rede
    def arq_rede(self):
        n_x = self.x.shape[0]
        n_h = self.hidden_neurons
        n_y = self.y.shape[0]
        arquite = {'n_x': n_x, 'n_h': n_h, 'n_y': n_y}
        return(arquite)

#Inicializar pesos
    def init_pesos(self):
        n_x = self.arq_rede().get('n_x')
        n_h = self.arq_rede().get('n_h')
        n_y = self.arq_rede().get('n_y')
        W1 = np.full((n_h, n_x), np.random.uniform(0,n_h * n_x))*0.01
        W2 = np.full((n_y, n_h), np.random.uniform(0,n_h * n_x))*0.01
        params_pesos = {'W1': W1, 'W2': W2}
        return(params_pesos)
    

#Forward Propagation

    def forward_prop(self):
    
        n_h = self.arq_rede().get('n_x')
        n_y = self.arq_rede().get('n_y')
    
        W1 = self.init_pesos().get('W1')
        W2 = self.init_pesos().get('W2')
    
    
        Z1 = np.dot(W1,self.x) #Multiplicação entre camada de entrada X Pesos
        A1 = sig(Z1) #Função Sigmoide do resultado da Multiplicação entre Camada de entrada X Pesos (Camada Oculta)
        Z2 = np.dot(W2,A1) #Multiplicação entre o resultado da sigmoide X pesos da camada intermediaria (Entre camada oculta e saida)
        prev1 = sig(Z2) #Função Sigmoide no resultado da multiplicação de Z2 (Camada de saida)
        cache = {'Z1':W1, 'A1':A1, 'Z2':Z2, 'prev1':prev1}
        return (cache)
    
#Cost Function

    def funcao_custo(self):
    
        m = self.y.shape[1]
        prev1 = self.forward_prop().get('prev1')
        
        cost = sum((self.y - self.forward_prop().get('prev1'))**2)/m
        return (cost)
    
#Backpropagation
    def backwardPropagation(self):
    
        m = self.y.shape[1]
        
        n_x = self.arq_rede().get('n_x')
        n_h = self.arq_rede().get('n_h')
        n_y = self.arq_rede().get('n_y')

        prev1 = self.forward_prop().get('prev1')
        A1 = self.forward_prop().get('A1')
        W2 = self.init_pesos().get('W2')


        dZ2 = self.forward_prop().get('prev1') - self.y #Valor previsto - Valor real (Diferença para alterar os pesos)
        dW2 = 1/m * np.dot(dZ2, A1.T) #Quanto eu preciso alterar os pesos de W2


        dZ1 = np.dot(W2.T,dZ2) * (1 - A1**2)#Mudança que ocorre na camada intermediaria para camada de entrada em funçao de DZ2
        dW1 = 1/m * np.dot(dZ1 , self.x.T) #Quanto eu preciso alterar os pesos de W1 em função de W2


        grads = {'dW1':dW1, 'dW2':dW2}


        return(grads)
    
#Atualizar os pesos
    def updateParameters(self):
    
        W1 = self.init_pesos().get('W1')
        W2 = self.init_pesos().get('W2')

        dW1 = self.backwardPropagation().get('dW1')
        dW2 = self.backwardPropagation().get('dW2')


        W1 = W1 - self.lr * dW1 #Atualizar peso W1
        W2 = W2 - self.lr * dW2 #Atualizar peso W2

        updated_pesos = {'W1': W1, 'W2': W2}


        return (updated_pesos)
    



```
