```python
import numpy as np
import pandas as pd
import string
import warnings
warnings.filterwarnings('ignore')

import requests
```

```python
url = 'https://www.soccerstats.com/matches.asp?matchday=2&daym=tomorrow'
```

```python
header = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"  
 }

r = requests.get(url, headers=header)
```

```python
#Verificar em qual ponto começa a tabela
pd.read_html(r.text,  decimal='.')[6]
```

```python
#Loop para percorrer as tabelas do site e juntar em 1 só dataframe
p = 7
k = [] 
j = pd.read_html(r.text,  decimal='.')[6]#Ponto do inicio da tabela
u=0
for i in range(1,9):
    g = pd.read_html(r.text,  decimal='.')[p]
    k.append(g)
    a = pd.DataFrame(k[u])
    j = pd.concat([j,a])
    u+=1
    p+=1
    
```

```python
#filtrar as colunas importantes e Mudar nome das colunas
j = j.iloc[:,0:14]
j.columns = ['Time','Hora','Mando','GP','W','FTS','CS','BTS','TG','GF','GA','1.5','2.5','3.5']
j
```

```python
#Filtrar o times que jogam em casa e os times que jogam fora de casa
hom = j[j['Mando']=='home']
aw = j[j['Mando']=='away']
```

```python
#Excluir as colunas Hora e Mando
hom = hom.drop(['Hora','Mando'],axis=1)
aw = aw.drop(['Hora','Mando'],axis=1)
```

```python
#Resetar o index para juntar os dataframes (Casa e Fora)
hom1 = hom.reset_index()
aw1 = aw.reset_index()

#Exluir coluna index
hom1 = hom1.iloc[:,1:]
aw1 = aw1.iloc[:,1:]


#Alterar nome das variáveis
aw1.columns = ['Time1','GP1','W1','FTS1','CS1','BTS1','TG1','GF1','GA1','1.51','2.51','3.51']
aw1 = aw1.iloc[:,1:]
aw1
```

```python
#Juntar os datasets (Casa e Fora)
data_new = hom1.join(aw1)

#Filtrar somente os numeros
data_new = data_new.loc[(data_new.GP!='(no home result recorded yet)')&(data_new.GP1!='(no away result recorded yet)')]
data_new.head()
```

```python
#Deletar o Caracter especial % da tabela
def corrigir_nomes(nome):
    nome = nome.replace('%', '')
    return nome

p = 1
for i in range(1,23):
    data_new.iloc[:,p] =data_new.iloc[:,p].apply(corrigir_nomes)
    p+=1
data_new
```

```python
#transformar os dados em numericos
rr = data_new.iloc[:,1:]
rr = rr.astype(float)

#Juntar os dados novamente
name = pd.DataFrame(data_new.iloc[:,0])
dados = name.join(rr)

#Filtrar os jogos que tem partidas >3

dados = dados.loc[(dados['GP']>4)&(dados['GP1']>4)]
dados
```

```python
#Excluir a colunas do numero de partidas ("GP","GP1")
dados = dados.drop(["GP","GP1"],axis=1)
dados
```

```python
#Exportar planilha excel
dados.to_csv("01-06-Total.csv")
```
