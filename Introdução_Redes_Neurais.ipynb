$$Conceitos:$$

- **Camadas de entrada não pode ser menor do que a quantidade de variáveis que estamos usando para explicar Y**


- **Deep Learning** = Arquitetura da Rede Neural


- **Redes neurais rasas (shallow)** = Possui apenas uma camada oculta


- **Redes neurais profundas** = Quando as redes neurais possuem muitas camadas ocultas, trata-se de deep learning.


- **Função de ativação** = 


- **Função de Custo** = Função que mensura o quanto estamos acertando


- **Descida do gradiente** = Processo utilizado para encontrar o mínimo da função custo com base em variação nos pesos, através do processo de backpropagation os erros obtidos após os cálculos dos pesos são retroalimentados na rede até otimizarmos o resultado


**Tipos de Modelos de Redes Neurais:**


- **Perceptron (P), Feed Forward Network (FFN), Radial Basis Network (RBF)** = São o tipo mais básico, onde a informação de entrada flui em sequência linear até a saída.


```python
#Importar as bibliotecas
import numpy as np
import pandas as pd
```

## Carregando o dataset

```python
#Carregar dataset
dados = pd.read_csv('winequality-red.csv',encoding='latin1',sep=',')
dados
```

```python
#Separar os dados em X e Y
x = dados.iloc[:,0:11]
y = dados.quality

#Padronização dos dados

dados_padro = (x - x.mean()) / (x.std())
dados_padro
```

```python
#Divisão dos dados 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dados_padro,y,test_size=0.3,random_state=1)

#Alterar as dimensoes
X_train = X_train.T
X_test = X_test.T
y_train = pd.DataFrame(y_train).T
y_test = pd.DataFrame(y_test).T
```

# Parte 1 - Criar arquitetura da rede

```python
#Arquitetura
def arq_func(x,y,hidden_neurons):
    n_x = x.shape[0] #quantidade de linhas de x = neurônios da camada de entrada
    n_h = hidden_neurons #quantidade de neurônios na camada escondida
    n_y = y.shape[0] #quantidade de linhas de y = neurônios da camada de saída
    arquite = {'n_x': n_x, 'n_h': n_h, 'n_y': n_y}
    return(arquite)
    
    
#Aplicar a função arquitetura de rede
arq_rede = arq_func(X_train, y_train , 4)
arq_rede
```

# Parte 2 - Inicializa Parâmetros randomicamente

- A partir da arquitetura, inicializamos os parâmetros randomicamente. O primeiro conjunto é W1 e b1. O segundo é  W2 and b2.O valor destes parâmetros dependem do tamanho das camadas de entrada e saída.

```python
#Inicializar pesos
from random import randrange, uniform

def init_pesos(x, arq_rede):
    n_x = arq_rede.get('n_x')
    n_h = arq_rede.get('n_h')
    n_y = arq_rede.get('n_y')
    
    W1 = np.full((n_h, n_x), np.random.uniform(0,n_h * n_x))*0.01
    W2 = np.full((n_y, n_h), np.random.uniform(0,n_h * n_x))*0.01
    
    params_pesos = {'W1': W1, 'W2': W2}
    return(params_pesos)
    
    
    
inic_pesos = init_pesos(x,arq_rede)
inic_pesos
```

# Parte 3 - Função de ativação - Sigmoide

```python
def sig(x):
    return((1/(1+np.exp(-x))))
```

# Parte 4 - Forward Propagation

- Agora com todas as informações dos pesos e estrutura da rede, iniciamos o processo de geração de resultados, multiplicação das matrizes (Variáveis e pesos)

```python
def forward_prop(x,params_pesos,arq_rede):
    
    n_h = arq_rede.get('n_h')
    n_y = arq_rede.get('n_y')
    
    W1 = inic_pesos.get('W1')
    W2 = inic_pesos.get('W2')
    
    
    Z1 = np.dot(W1,x)
    A1 = sig(Z1)
    Z2 = np.dot(W2,A1)
    A2 = sig(Z2)
    
    cache = {'Z1':W1, 'A1':A1, 'Z2':Z2, 'A2':A2}

    return (cache)
    
```
